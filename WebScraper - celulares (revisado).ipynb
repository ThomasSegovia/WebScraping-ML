```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 

url = 'https://www.mercadolibre.com.ar/ofertas?domain_id=MLA-CELLPHONES#origin=scut&filter_applied=domain_id&filter_position=3&is_recommended_domain=false'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

#precios actuales celulares

prcel = soup.find_all('span', class_="promotion-item__price")

precioActual = list()

count = 0
for i in prcel:
    if count < 40:
        precioActual.append(i.text)
    else:
        break
    count +=1    

#print(precioActual)

#precios viejos celulares

preciosAnteriores= soup.find_all('span', class_="promotion-item__oldprice")

precioAnt= list()

count = 0
for i in preciosAnteriores:
    if count < 40:
        precioAnt.append(i.text)
    else:
        break
    count +=1    

    ############################################################

descuento= soup.find_all('span', class_="promotion-item__discount")

preciooff= list()

count = 0
for i in descuento:
    if count < 40:
        preciooff.append(i.text)
    else:
        break
    count +=1    

############################################################

envioDisponible= soup.find_all('span', class_="promotion-item__shipping")

envio= list()

count = 0
for i in envioDisponible:
    if count < 40:
        envio.append(i.text)
    else:
        break
    count +=1
    
############################################################

descripcionDeObjeto= soup.find_all('p', class_="promotion-item__title")

descripcion= list()

count = 0
for i in descripcionDeObjeto:
    if count < 40:
        descripcion.append(i.text)
    else:
        break
    count +=1    
```

```python
df = pd.DataFrame({'Precio':precioActual, 'PrecioAntiguo':precioAnt, 'Descuento':preciooff, 'Envio':envio, 'Descripcion':descripcion}, index=list(range(1,41)))
print(df)
```

```python
df.to_csv('precio_celulares_de_oferta.csv', index=False)
```
